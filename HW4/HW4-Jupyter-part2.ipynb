```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InLineBackend.figure_format = 'svg'
plt.style.use('fivethirtyeight')
```

```python
ccfd = pd.read_csv('./creditcard.csv', sep = ',')
ccfd
```

```python
data = ccfd.value_counts('Class')
data
```

```python
ccfd['Class'].value_counts()
```

```python
data.plot(kind='bar')
plt.show()
```

```python
data.plot(kind='bar', logy=True)
plt.show()
```

```python
a = ccfd[ccfd['Class'] == 0]['V1']
a
```

```python
b = ccfd[ccfd['Class'] == 1]['V1']
b
```

```python
plt.hist(a, bins=20, alpha=0.5, ec='white', color='grey', density='True')
plt.hist(b, bins=20, alpha=0.5, ec='white', color='red', density='True')
legend = plt.legend(labels=['Class 0', 'Class 1'], loc='upper left')
plt.xlabel('V1')
plt.ylabel('Density')
plt.show()
```
