```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
```

### Форматирование текста для графика

```python
x = np.linspace(-5, 5, 101)
x
```

```python
x2 = x**2
x2
```

```python
plt.plot(x, x2)
plt.show()
```

Добавим название графика:

```python
plt.plot(x, x2)
plt.title('График f(x)')
plt.show()
```

Поменяем настройки для большей четкости графики

```python
# SVG (Scalable Vector Graphics) - масштабируемая векторная графика
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(x, x2)
plt.title('График f(x)')
plt.show()
```

Изменение размера текста

```python
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16)
plt.show()
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16, fontweight='bold')
plt.show()
```

```python
# Указание названия цвета из таблицы HTML цветов
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16, fontweight='bold', color='DarkRed')
plt.show()
```

```python
# название цвета в формате HEX (шестнадцатиричная система исчисления)
plt.plot(x, x2)
plt.title('График f(x)', fontsize=16, fontweight='bold', color='#808080')
plt.show()
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'TimesNewRoman'}
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontdict=title_dict)
plt.show()
```

```python
plt.plot(x, x2)
# loc - расположение названия графика (по умолчанию - 'center')
plt.title('График f(x)', fontdict=title_dict, loc='left')
plt.show()
```

Добавление названия осей

```python
plt.plot(x, x2)

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x')
plt.ylabel('Функция f(x)')

plt.show()
```

```python
label_dict = {'fontsize': '14', 'color': '#808080', 'family': 'Calibri'}

plt.plot(x, x2)

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.show()
```

Добавим график ещё одной функции

```python
x3 =x**3
x3
```

```python
plt.plot(x, x2)
plt.plot(x, x3)

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.show()
```

### Ограничение осей

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2)
plt.plot(x, x3)

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.show()
```

### Добавление легенды

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2)
plt.plot(x, x3)

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.legend(labels=['x2', 'x3'])

plt.show()
```

2-й способ:

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.legend()

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.legend(loc='lower right')

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.legend(loc='lower right', fontsize=8)

plt.show()
```

```python
legend_dict = {'family': 'Calibri', 'size': 10}
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

plt.legend(loc='lower right', prop=legend_dict)

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График f(x)', fontdict=title_dict)

plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

legend = plt.legend(loc='lower right', prop=legend_dict)
plt.setp(legend.get_texts(), color='grey')

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

legend = plt.legend(loc='lower right', prop=legend_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2', color='#3bb08f')
plt.plot(x, x3, label='x**3', color='#e97451')

plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

legend = plt.legend(loc='lower right', prop=legend_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')
plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

plt.plot(x, x2, label='x**2', color='#3bb08f', linestyle='dashed')
plt.plot(x, x3, label='x**3', color='#e97451', linestyle='dotted')

plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная x', fontdict=label_dict)
plt.ylabel('Функция f(x)', fontdict=label_dict)

legend = plt.legend(loc='lower right', prop=legend_dict, frameon=False)
plt.setp(legend.get_texts(), color='grey')
plt.show()
```
