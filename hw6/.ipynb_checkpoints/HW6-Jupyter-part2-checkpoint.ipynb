```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
ccfd = pd.read_csv('./creditcard.csv', sep=',')
```

```python
ccfd
```

```python
ccfd['Class'].value_counts(normalize=True)
```

```python
ccfd.info()
```

```python
pd.options.display.max_columns = 100
```

```python
ccfd.head(10)
```

```python
X = ccfd.drop('Class', axis=1)
X
```

```python
y = ccfd['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]
```

```python
gscv_m = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                     param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
gscv_m.fit(X_train, y_train)
```

```python
gscv_m.best_params_
```

```python
y_pred_proba = gscv_m.predict_proba(X_test)
```

```python
y_pred_proba
```

```python
y_pred_proba = y_pred_proba[:, 1]
```

```python
y_pred_proba
```

```python
plt.hist(y_pred_proba[y_test == 1], bins = 7)
plt.xlabel('Probability')
plt.ylabel('Количество')
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
gscv_m.best_score_
```

Ответ: результат, полученный на тренеровачных данных несколько выше,
       чем результат, полученный на тестовых данных (0.97 > 0.95)
