```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
type(wine)
```

```python
data_keys = wine.keys()
data_keys
```

```python
data = wine.data
data
```

```python
features = wine.feature_names
features
```

```python
for line in wine.DESCR.split('\n'):
    print(line)
```

```python
target = wine.target
target
```

```python
np.unique(target)
```

```python
X = pd.DataFrame(data, columns=features)
X
```

```python
X.info()
```

```python
X['target'] = target.astype(int)
X
```

```python
X.info()
```

```python
X['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_corr = X.corr()
```

```python
plt.figure(figsize=(15, 10))
sns.set(font_scale=1.4)
sns.heatmap(X_corr, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

```python
X_corr
```

```python
high_corr = X_corr['target'][(np.abs(X_corr['target']) > 0.5) & (X_corr['target'] != 1)]
```

```python
high_corr
```

```python
high_corr = list(high_corr.keys())
high_corr
```

```python
X
```

```python
X = X.drop('target', axis=1)
```

```python
X
```

```python
for el in high_corr:
    el2 = el + '_2'
    X[el2] = X[el] ** 2
X  
```
